- 평균적으로 Walt Disney Studio Motion Pictures와 Sony Pictures가 높은 예산을 사용하는 것을 확인할 수 있다. 
- 그렇다면 매출은 어떠할까? (매출은 worldwide 즉 전세계 매출을 기준으로 한다)

```python
data=pd.concat([df['Worldwide'],df['Distributor']],axis=1)
f,ax=plt.subplots(figsize=(10,8))
fig=sns.boxplot(x='Distributor',y='Worldwide',data=data)
plt.xticks(rotation=60)
```

### 적은 리뷰 (각 영화당 50페이지)

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
import re

pres=['https://movie.naver.com/movie/bi/mi/pointWriteFormList.naver?code=72054&type=after&onlyActualPointYn=N&onlySpoilerPointYn=N&order=sympathyScore&page=',
     'https://movie.naver.com/movie/bi/mi/pointWriteFormList.naver?code=72054&type=after&onlyActualPointYn=N&onlySpoilerPointYn=N&order=lowest&page=']

review=[]
rate=[]
target=[]

for pre in pres:
    for i in range(1,50):
        url=pre+str(i)
        res=requests.get(url)
        soup=BeautifulSoup(res.content,'html.parser')
    
        id_list=[]
        id_pre='_filtered_ment_'
    
        for i in range(10):
            id_list.append(id_pre+str(i))
        
        for id in id_list:
            review.append(soup.find('span',{'id':id}).get_text().strip())
            
        rate_list=[]
        rate_list=(soup.select('div.star_score > em'))
        
        for i in range(10):
            r=int(re.sub('<.+?>','',str(rate_list[i])))
            rate.append(r)
            if r >= 8:
                target.append(1)
            elif r<=4:
                target.append(0)
            else:
                target.append(-1)
                
df=pd.DataFrame({'review':review,'rate':rate,'target':target})
```

```python
docs=df['review']
```

```python
N=len(docs) # 총 문서의 수

def tf(t,d):
    return d.count(t)

def idf(t):
    df=0
    for doc in docs:
        df+=t in doc
    return log(N/(df+1))

def tfidf(t,d):
    return tf(t,d) * idf(t)
```

```python
result=[]
for i in range(N): # 각 문서에 대해 아래 명령을 수행
    result.append([])
    d=docs[i]
    for j in range(len(noun_list)):
        t=noun_list[j]
        result[-1].append(tf(t,d))
        
tf_=pd.DataFrame(result,columns=noun_list)
tf_
```

```python
from numpy import dot
from numpy.linalg import norm
import numpy as np
def cos_sim(A,B):
    return dot(A,B)/(norm(A)*norm(B))
```

```python
doc1=np.array([3,4])
doc2=np.array([-1,2])
```

```python
print(cos_sim(doc1,doc2)) # 문서 1, 2의 코사인 유사도
```
