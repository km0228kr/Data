```python
import re

def password(pw):
    a=re.compile('[a-z]+')
    b=re.compile('[A-Z]+')
    c=re.compile('[0-9]+')
    d=re.compile('[#@&]+')
    
    if (
        (len(pw)>=8 and len(pw)<13)
        and len(a.findall(pw))>0
        and len(b.findall(pw))>0
        and len(c.findall(pw))>0
        and len(d.findall(pw))>0
    ):
        print('')
    else:
        print('')
        
        
password('abc12345')
```

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
import re

pres=['https://movie.naver.com/movie/bi/mi/pointWriteFormList.naver?code=72054&type=after&onlyActualPointYn=N&onlySpoilerPointYn=N&order=sympathyScore&page=',
     'https://movie.naver.com/movie/bi/mi/pointWriteFormList.naver?code=72054&type=after&onlyActualPointYn=N&onlySpoilerPointYn=N&order=lowest&page=']

review=[]
rate=[]
target=[]

for pre in pres:
    for i in range(1,50):
        url=pre+str(i)
        res=requests.get(url)
        soup=BeautifulSoup(res.content,'html.parser')
    
        id_list=[]
        id_pre='_filtered_ment_'
    
        for i in range(10):
            id_list.append(id_pre+str(i))
        
        for id in id_list:
            review.append(soup.find('span',{'id':id}).get_text().strip())
            
        rate_list=[]
        rate_list=(soup.select('div.star_score > em'))
        
        for i in range(10):
            r=int(re.sub('<.+?>','',str(rate_list[i])))
            rate.append(r)
            if r >= 8:
                target.append(1)
            elif r<=4:
                target.append(0)
            else:
                target.append(-1)
                
df=pd.DataFrame({'review':review,'rate':rate,'target':target})
```

```python
import pandas as pd
from math import log # IDF  
#  
from konlpy.tag import Okt
from collections import Counter
```

```python
okt=Okt() #  
word_list=[]
word_list=df['review']
sentences_tag=[]

for sentence in word_list:
    morph=okt.pos(sentence)
    sentences_tag.append(morph)
```

```python
sentences_tag
```

```python
noun_list=[]
for sentence in sentences_tag:
    for word,tag in sentence:
        if tag in ['Noun']:
            noun_list.append(word)
```

```python
noun_list=[n for n in noun_list if len(n) > 1]
noun_list[:100]
```

```python
noun_list=list(set(noun_list))
noun_list.sort()
```

```python
noun_list
```

```python
docs=df['review']
```

```python
N=len(docs) #   

def tf(t,d):
    return d.count(t)

def idf(t):
    df=0
    for doc in docs:
        df+=t in doc
    return log(N/(df+1))

def tfidf(t,d):
    return tf(t,d) * idf(t)
```

```python
result=[]
for i in range(N): #      
    result.append([])
    d=docs[i]
    for j in range(len(noun_list)):
        t=noun_list[j]
        result[-1].append(tf(t,d))
        
tf_=pd.DataFrame(result,columns=noun_list)
tf_
```

```python
result=[]
for j in range(len(noun_list)):
    t=noun_list[j]
    result.append(idf(t))
    
idf_=pd.DataFrame(result,index=noun_list,columns=['IDF'])
idf_
```

```python
result=[]
for i in range(N):
    result.append([])
    d=docs[i]
    for j in range(len(noun_list)):
        t=noun_list[j]
        
        result[-1].append(tfidf(t,d))
        
tfidf_=pd.DataFrame(result,columns=noun_list)
tfidf_
```
